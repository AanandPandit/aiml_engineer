```python
# Reshaping of arrays

arr = np.array([1,2,3,4,5,6])
print(arr)
print('------------')
reshaped_arr = arr.reshape(2,3)   # (rows, col)
print(reshaped_arr)
```

```python
# ravel array converts multi dimensional array to single dimensional array

changed_arr = reshaped_arr.ravel()
print(changed_arr)
```

```python
# flatten array also converts multi dimensional array to single dimensional array

flat_arr = reshaped_arr.flatten()
print(flat_arr)
```

```python
#Views and copies

arr = np.array([1,2,3,4,5,6])
print(arr)

#  Views change in view reflects in original array
arr[1] = 100
print(arr)
```

```python
# Copies do not change in original array
arr2 = arr.copy()
print(arr2)

arr2[2] = 300 
print(arr2) # copy changed
print(arr)  # original array remains unchanged
```

```python
# Aggregate functions in numpy


arr = np.array([1,2,3,8,5,6,4,7])
print(arr.sum())
print(arr.min( ))
print(arr.max())
print(arr.mean())
print(arr.std())   # standard deviation
print(arr.var())  # variance
print(np.median(arr))  # median
print(np.percentile(arr, 50))  # 50th percentile
print(np.percentile(arr, 90))  # 90th percentile
print(np.percentile(arr, 25))  # 25th percentile
print(np.percentile(arr, 75))  # 75th percentile
print(np.argmin(arr))  # index of minimum value
print(np.argmax(arr))  # index of maximum value
print(np.argsort(arr))  # indices that would sort the array
print(np.unique(arr))  # unique elements in the array
```
