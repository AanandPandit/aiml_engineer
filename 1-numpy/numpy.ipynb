```python
# multi dimensional array
```

```python
arr = np.array([[1,2,3,4],
               [4,5,6,7]],
               dtype=np.float32)
print(arr)
```

```python
print(arr.ndim)
print(arr.shape)
print(arr.dtype)
print(arr.itemsize)   # size of each element in bytes
print(arr.size)       # total number of elements
```

```python
# Reshaping of arrays

arr = np.array([1,2,3,4,5,6])
print(arr)
print('------------')
reshaped_arr = arr.reshape(2,3)   # (rows, col)
print(reshaped_arr)
```

```python
# ravel array converts multi dimensional array to single dimensional array

changed_arr = reshaped_arr.ravel()
print(changed_arr)
```

```python
# flatten array also converts multi dimensional array to single dimensional array

flat_arr = reshaped_arr.flatten()
print(flat_arr)
```

```python
# Array Indexing (Advanced Indexing)
```

```python
arr = np.array([34,5,6,34,5,6,34,5,6,34,5,6,34,5,6,34,5,6])
indexes = [0,3,5,9]   # we want to access these indexes

arr.take(indexes)
```

```python
arr = np.array([[21,22,23,24,25],
                [34,35,36,37,38]])

for i in np.nditer(arr):
    print(i, end=' ')
```

```python
for (value, index) in np.ndenumerate(arr):
    print(index, value)
```
