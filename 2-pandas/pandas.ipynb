```python
df.dropna(inplace=True)
df
```

```python
# add more 10 data to dataset with default values and no null value  to extend the dataframe and add to original dataframe
new_data = {
    'full_name': ['Eman e', 'Fanan f', 'Ganan g', 'Hanan h', 'Iman i', 'Janan j', 'Kanan k', 'Lanan l', 'Manan m', 'Manan m'],
    'age_in_years': [23, 24, 25, 26, 27, 28, 29, 30, 31, 31],
    'department': ['CSE', 'ECE', 'MECH', 'CIVIL', 'CSE', 'ECE', 'MECH', 'CIVIL', 'CSE', 'CSE'],
    'salary': [70000, 72000, 68000, 75000, 71000, 73000, 69000, 76000, 74000, 74000],
    'promoted_salary': [98000.0, 100800.0, 95200.0, 105000.0, 99400.0, 102200.0, 96600.0, 106400.0, 103600.0, 103600.0]
}
new_df = pd.DataFrame(new_data)
new_df  
df = pd.concat([df, new_df], ignore_index=True)
df
```

```python
df.info()
```

```python
df.dropna(how='any')    # drop rows with any null values
df.dropna(how='all')   # drop rows with all null values
```

```python
# fill nas with mean value of respective column
df['age_in_years'] = df['age_in_years'].fillna(df['age_in_years'].mean())
df['salary'] = df['salary'].fillna(df['salary'].mean())
df['promoted_salary'] = df['promoted_salary'].fillna(df['promoted_salary'].mean())
df
```

```python
df['age_in_years'].fillna(method='ffill')   # forward fill
df['age_in_years'].fillna(method='bfill')   # backward fill
df
```

```python
df['age_in_years'].bfill()
df
```

```python
df['full_name'].replace({
    'Anan': 'Aanand'
}, inplace=True)
df
```

```python
# find the duplicate rows
df_dup = df[df.duplicated()]
df_dup
```

```python
# drop duplicates and make changes in original dataframe
df.drop_duplicates(inplace=True)
df
```

```python
# split first and last names into two different columns
df[['first_name', 'last_name']] = df['full_name'].str.split(' ', expand=True, n=1)
df
```

```python
location_data = {
    'location': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose', 'Austin', 'Jacksonville', 'Fort Worth', 'Columbus'],
    'department': ['CSE', 'ECE', 'MECH', 'CIVIL', 'CSE', 'ECE', 'MECH', 'CIVIL', 'CSE', 'CSE', 'CSE', 'ECE', 'MECH', 'CIVIL']
}
df3 = pd.DataFrame(location_data)
df3
```

```python
# concat df and df3 on department column
df = pd.concat([df, df3], axis=1)
df
```

```python
# merge  df and df3 on department column, keeping only matching rows

df = pd.merge(df, df3, on='department', how='inner')
df
```

```python
df.columns.tolist()
```

```python
# keep the first occurrence of each column name, drop later duplicates and update in original dataframe
df = df.loc[:, ~df.columns.duplicated()]
# verify
print(df.columns.tolist())
df.head()
```
