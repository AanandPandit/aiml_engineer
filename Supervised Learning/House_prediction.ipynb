## Dataset Overview

The dataset contains real estate features with the following attributes:

| Feature          | Description                             |
|------------------|-----------------------------------------|
| **MedInc**       | Median income in the area               |
| **HouseAge**     | Median house age                        |
| **AveRooms**     | Average number of rooms                 |
| **AveBedrms**    | Average number of bedrooms              |
| **Population**   | Population of the area                  |
| **AveOccup**     | Average number of occupants             |
| **Latitude**     | Latitude of the area                    |
| **Longitude**    | Longitude of the area                   |
| **MedHouseVal** (Target) | House price (Target variable)      |


```python
# Step-1
# import libraties and load data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

from sklearn.datasets import fetch_california_housing

data = fetch_california_housing(as_frame=True)
df = data.frame

df.head()
```

```python
# Step-2
# Basic data cleaning

df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# the dataset has no missing values and all features are numerical. 
```

```python
# Step-3 
# Exploratory Data Analysis

# 1. Pairplot of important features 
# Pairplot to visualize relationships between selected features

sns.pairplot(df[['MedInc', 'HouseAge', 'AveBedrms', 'MedHouseVal']])
plt.show()
```

```python
# 2. Correlation Heatmap
# Heatmap to visualize correlation between features which means how features are related to each other 

plt.figure(figsize=(10, 7))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.show()
```

```python
# 3. Check the target variable distribution which means how the target variable is distributed

sns.histplot(df['MedHouseVal'], kde=True)
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```
